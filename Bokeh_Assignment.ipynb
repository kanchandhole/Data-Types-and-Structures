**4.**Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Spectral11 as colors
from bokeh.models import ColumnDataSource

# Generate the data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource to use with Bokeh
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

# Create a figure object
p = figure(title="Histogram of Data Distribution", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Create the histogram bars
p.quad(top='top', bottom=0, left='left', right='right', source=source,
       fill_color=colors[0], line_color="white", alpha=0.7)

# Show the plot
output_notebook()
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256 as colors
from bokeh.layouts import column

# Generate the dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the meshgrid and dataset
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create DataFrame for the heatmap
df = pd.DataFrame({'x': x_flat, 'y': y_flat, 'z': z_flat})

# Define color map
color_mapper = linear_cmap(field_name='z', palette=colors, low=min(z_flat), high=max(z_flat))

# Create a figure object
p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y', width=600, height=600)

# Add square glyphs for the heatmap
p.square(x='x', y='y', size=20, color=color_mapper, source=df)

# Add a color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
