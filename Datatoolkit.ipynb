**5.**Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from array

```python
# Create a 5x6 array with random integers between 1 and 10
arr = np.random.randint(1, 11, size=(5, 6))

# a) Extract all even integers from the array
even_integers = arr[arr % 2 == 0]

# b) Extract all odd integers from the array
odd_integers = arr[arr % 2 != 0]

print("Original Array:")
print(arr)

print("\nEven integers:")
print(even_integers)

print("\nOdd integers:")
print(odd_integers)
```

**17.**Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
# Generate two random arrays for x and y coordinates
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame with x, y and quadrant information
data = pd.DataFrame({
    'x': x,
    'y': y
})

# Classify points based on which quadrant they lie in
def get_quadrant(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Quadrant 1'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Quadrant 2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

# Apply the function to classify points
data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create a Seaborn scatter plot, coloring points by quadrant
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='Set1', s=100)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

**19.**Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral11
from bokeh.transform import linear_cmap
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame to hold the categories and values
data = pd.DataFrame({
    'categories': categories,
    'values': values
})

# Convert to ColumnDataSource for Bokeh
source = ColumnDataSource(data)

# Create a new figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value', width=800, height=400)

# Color bars based on their values using a color mapper
color_mapper = linear_cmap('values', Spectral11, min(values), max(values))

# Generate the bar chart with colors based on values
p.vbar(x='categories', top='values', source=source, width=0.6,
       color=color_mapper, legend_field='categories')

# Add hover tool to display the exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Set axis labels and title
p.title.text_font_size = '16pt'
p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'

# Display the plot
show(p)

```
